Here's a brief version of what you'll find in the data description file.

**SalePrice** - цена продажи недвижимости в долларах. Это целевая переменная, которую вы пытаетесь предсказать.

```python
def mod_outlier(df):
        df1 = df.copy()
        df = df._get_numeric_data()


        q1 = df.quantile(0.25)
        q3 = df.quantile(0.75)

        iqr = q3 - q1

        lower_bound = q1 -(1.5 * iqr) 
        upper_bound = q3 +(1.5 * iqr)


        for col in df.columns:
            for i in range(0,len(df[col])):
                if df[col][i] < lower_bound[col]:            
                    df[col][i] = lower_bound[col]

                if df[col][i] > upper_bound[col]:            
                    df[col][i] = upper_bound[col]    


        for col in df.columns:
            df1[col] = df[col]

        return(df1)

df_new_feat = mod_outlier(df_new_feat)
```

```python
@interact
def plot_tree_rf(crit=["mse", "entropy"],
              bootstrap=["True", "False"],
              depth=IntSlider(min=1, max=30, value=3, continuous_update=False),
              min_split=IntSlider(min=2, max=5, value=2, continuous_update=False),
              min_leaf=IntSlider(min=1, max=5, value=1, continuous_update=False),
              forests=IntSlider(min=1, max=1000, value=100, continuous_update=False)):
  dtr = RandomForestRegressor(random_state=1,
                                     criterion=crit,
                                     bootstrap=bootstrap,
                                     n_estimators=forests,
                                     max_depth=depth,
                                     min_samples_leaf=min_leaf,
                                     min_samples_split=min_split,
                                     n_jobs = -1,
                                     verbose=False
                                     )
  dtr.fit(x_train, y_train)
  print('Decision Tree Training score: {:.3f}'.format(dtr.score(x_train, y_train)))
  print('Decision Tree Testing score: {:.3f}'.format(dtr.score(x_test, y_test)))
  rmse = math.sqrt(mean_squared_error(y_test, rfr.predict(x_test)))
  print("mear squares error :",rmse)
  num_tree = dtr.estimators_[0]
  graph = Source(tree.export_graphviz(num_tree,out_file=None,
                                      feature_names = x_train.columns,
                                      class_names=['stay','left'],
                                      filled=True))
  display(Image(data=graph.pipe(format='png')))
  return dtr
```
